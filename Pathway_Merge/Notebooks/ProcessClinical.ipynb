```python
cancer = 'BRCA'
br = clin[cancer]
```

```python
cols = [c for c, vals in br.iterrows() if (sum(notnull(vals)) > len(vals)*.5)
        and len(set(vals.dropna())) > 1 and ('admin' not in c) and (c not in common)
        and ('patient.drugs.drug' not in c) and ('patient.followups.followup' not in c)]
```

```python
brca_cols = {'patient.ajcccancermetastasisstagecode': 'metastasis_stage',
 'patient.ajccneoplasmdiseaselymphnodestage': 'lymphnode_stage',
 'patient.ajccneoplasmdiseasestage': 'disease_stage',
 'patient.ajcctumorstagecode': 'tumor_stage',
 'patient.breastcarcinomaestrogenreceptorstatus': 'ER',
 'patient.breastcarcinomaprogesteronereceptorstatus': 'PR',
 'patient.clinicalcqcf.histologicalsubtype': 'histo_subtype',
 'patient.labprocedureher2neuinsituhybridoutcometype': 'her2_fish',
 'patient.labprocher2neuimmunohistochemistryreceptorstatus': 'her2_chem',
 'patient.lymphnodeexaminedcount': 'lymph_tested',
 'patient.menopausestatus': 'menopause',
 'patient.numberoflymphnodespositivebyhe': 'lymph_positive',
 'patient.personneoplasmcancerstatus': 'inital_tumor'}
```

```python
c = br.ix[brca_cols]
c = c.rename(index=brca_cols).T
c = c.rename(index=br.ix['patient.bcrpatientbarcode'])
c['menopause'] = c.menopause.dropna().apply(lambda s: s.split(' ')[0]) #get rid of description
```

```python
def her2_combo(s):
    if s['her2_fish'] in ['positive','negative']:
        return s['her2_fish']
    elif s['her2_chem'] in ['positive','negative']:
        return s['her2_chem']
    elif s['her2_fish'] not in [nan,'performed but not available','not performed']:
        return s['her2_fish']
    else:
        return s['her2_chem']
    
c['her2'] = c.apply(her2_combo, 1)
```

```python
stage_lo = {'stage i' : 1, 'stage ia' : 1, 'stage ib' : 1, 'stage ii' : 2, 
            'stage iia' : 2, 'stage iib' : 2, 'stage iii' : 3, 'stage iiia' : 3,
            'stage iiib' : 3, 'stage iiic' : 3, 'stage iv' : 4, 'stage tis' : nan,
             'stage x' : nan, nan : nan}
to_simple_stage = lambda s: stage_lo[s] if s.startswith('stage') else s[:2]
stages_simple = DataFrame(dict([(var, c[var].dropna().map(to_simple_stage)) for var in 
                               ['tumor_stage', 'disease_stage', 'lymphnode_stage', 
                                'metastasis_stage']]))
fig, axs = subplots(1,4, figsize=(12,2.5))
for i,(var, vals) in enumerate(stages_simple.iteritems()):
    counts = Series(dict((s, sum(vals == s)) for s in set(vals.dropna())))
    counts.plot(kind='bar', ax=axs[i])
    axs[i].set_title(var)
```

```python
followup = br.select(lambda s: s.startswith('patient.followups.followup'))
mIdx = map(lambda s: ('.'.join(s.split('.')[:4]), '.'.join(s.split('.')[4:])), 
           followup.index)
mIdx = MultiIndex.from_tuples(mIdx)
followup = DataFrame(followup.as_matrix(), index=mIdx, columns=br.ix['patient.bcrpatientbarcode'])
keep = map(lambda s: type(s) == str, followup.columns)
followup = followup.ix[:,keep]
followup = followup.dropna(how='all', axis=[0,1])
```

```python
cols = ['vitalstatus','newneoplasmeventoccurrenceanatomicsite','metastaticbreastcarcinomaestrogenreceptorstatus',
        'daystolastfollowup', 'metastaticbreastcarcinomaestrogenreceptorstatus', 'newneoplasmeventtype', 'her2immunohistochemistrylevelresult']
patient_followup = {}
for patient, d in followup.iteritems():
    df = d.unstack().dropna(how='all', axis=[0,1])
    to_date = lambda s: datetime(int(s[0]), int(s[1]), int(s[2])) if notnull(s[0]) else nan
    _date = df[['yearofformcompletion','monthofformcompletion', 'dayofformcompletion']]
    _date = Series(_date.apply(to_date, 1), name='dateofformcompletion')
    df = df.select(lambda s: 'formcompletion' not in s, 1).join(_date)
    patient_followup[patient] = df
```

```python
event = [(p, vals['newtumoreventafterinitialtreatment'].ix[-1]) for p,vals in patient_followup.iteritems() if ('newtumoreventafterinitialtreatment' in vals)]
event = Series(dict(event)).dropna()
event = Series(np.sort(event), name='new_tumor')

time_to_event = [(p, vals['daystonewtumoreventafterinitialtreatment'].ix[-1]) for p,vals in patient_followup.iteritems() if ('newtumoreventafterinitialtreatment' in vals) and 
                                                 (sum(vals['newtumoreventafterinitialtreatment'] == 'yes') > 0) and 'daystonewtumoreventafterinitialtreatment' in vals]
time_to_event = Series(dict(time_to_event), name='days_to_event').dropna().astype(int)
time_to_event = np.sort(time_to_event)
```

```python
meta = meta_data[cancer]
timeline = meta[['daystodeath', 'daystolastfollowup']]
timeline = timeline.join(time_to_event).astype(float)
event_free_survival = Series(timeline.min(1), name='event_free_survival')
timeline = timeline.join(event_free_survival).join(meta.daystobirth)
```

```python
cl = c[['ER','PR','her2','menopause']].join(stages_simple).join(timeline).join(event)
cl = cl.join(meta[['vitalstatus','gender','tissuesourcesite']])

cl['age']  = cl['daystobirth'].astype(float) / -365.
cl['days'] = cl[['daystolastfollowup','daystodeath']].astype(float).max(1)
cl['deceased'] = (cl.vitalstatus.dropna() == 'deceased').astype(int)
cl['new_tumor'] = (cl['new_tumor'].dropna() == 'yes').astype(int)
cl['event'] = (cl['deceased'] + cl['new_tumor']).clip_upper(1)
cl['her2_pos'] = cl.her2.dropna() == 'positive'
cl['PR_pos'] = cl.PR.dropna() == 'positive'
cl['ER_pos'] = cl.ER.dropna() == 'positive'
cl['triple_neg'] = (cl[['ER','PR','her2']] == 'negative').sum(1) == 3
cl['triple_pos'] = (cl[['ER','PR','her2']] == 'positive').sum(1) == 3
cl['lymphnode_n0n1'] = cl.lymphnode_stage.dropna().apply(lambda s: s in ['n0','n1'])
cl['tumor_t1t2'] = cl.tumor_stage.dropna().apply(lambda s: s in ['t1','t2'])
cl['post_menopause'] = cl.menopause.dropna() == 'post'
```

```python
cl = cl.rename(index=str.upper)
out_path = '/'.join([stddata_path, cancer, date,'Clinical']) + '/' + 'compiled.csv'
cl.to_csv(out_path)
```
