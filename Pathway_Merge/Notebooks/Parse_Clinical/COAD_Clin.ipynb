```python
import os
from pandas import DataFrame, read_table, notnull, MultiIndex, datetime
```

```python
firehose_path = '/cellar/users/agross/Data/TCGA/'
date = '20120913'
date_ = '_'.join([date[:4], date[4:6], date[6:8]])
stddata_path = firehose_path + 'stddata__' + date_
cancers = os.listdir(stddata_path)
```

```python
clin_file = lambda c: '/'.join([stddata_path, c, date,'Clinical', c]) + '.clin.merged.txt' 
```

```python
clin = dict((c,read_table(clin_file(c), index_col=0)) for c in cancers
             if os.path.isfile(clin_file(c)))
common = set.intersection(*map(lambda v: set(v.index), clin.values()))
```

```python
keeper_columns = ['bcrpatientbarcode', 'gender', 'daystoinitialpathologicdiagnosis',
                  'daystobirth', 'race', 'icdo3histology', 'vitalstatus', 
                  'patientid', 'tissuesourcesite','daystodeath',
                  'ageatinitialpathologicdiagnosis', 'daystolastfollowup',
                  'ethnicity']
keeper_columns = map(lambda s: 'patient.' + s, keeper_columns)
```

```python
from pandas import Panel, DataFrame, Series
meta_data = {}
```

```python
for c,br in clin.iteritems():
    meta = br.ix[keeper_columns]
    meta = DataFrame(meta.T.as_matrix(), columns=map(lambda s: s.split('.')[-1], 
                     meta.index), index=meta.ix['patient.bcrpatientbarcode'])
    #meta = meta.dropna(how='all', axis=[0,1])
    assert sum(meta['daystoinitialpathologicdiagnosis'] == 0) == 0
    meta_data[c] = meta
```

```python
figsize(11,5)
num_dead = DataFrame(dict((c, meta.groupby('gender').count().ix[:,0]) for 
                  c,meta in meta_data.iteritems())).T
num_dead.plot(kind='bar', stacked=True)
ylabel('Number of Patients')
```

```python
figsize(11,5)
num_dead = DataFrame(dict((c, meta.groupby('race').count().ix[:,0]) for 
                  c,meta in meta_data.iteritems())).T
num_dead.plot(kind='bar', stacked=True)
ylabel('Number of Patients')
```

```python
c  = DataFrame(index=br.columns)
c['metastasis'] = br.ix['patient.distantmetastasispathologicspread']
c['hist_type'] = br.ix['patient.histologicaltype']
c['colon_polyps'] = br.ix['patient.lossexpressionofmismatchrepairproteinsbyihc']
c['lymphatic_invasion'] = br.ix['patient.lymphaticinvasion']
c['lymphnode_stage'] = br.ix['patient.lymphnodepathologicspread']
c['lymphnode_n0n1']= c.lymphnode_stage.map(lambda s: ('n0' in s) or ('n1' in s), 
                                           na_action='ignore')
c['relatives'] = br.ix['patient.numberoffirstdegreerelativeswithcancerdiagnosis'] != '0'
c['tumor_grade'] = br.ix['patient.primarytumorpathologicspread']
c['tumor_t0t1']= c.tumor_grade.map(lambda s: ('t0' in s) or ('t1' in s), 
                                           na_action='ignore')
c['tumor_stage'] = br.ix['patient.tumorstage'].map(lambda s: 
                                  s.replace('a','').replace('b','').replace('c',''),
                                  na_action='ignore')
c['tissue'] = br.ix['patient.tumortissuesite']
c['venous_invasion'] = br.ix['patient.venousinvasion']
c = c.rename(index=br.ix['patient.bcrpatientbarcode'])
```
