```python
import os
from pandas import DataFrame, read_table, notnull, MultiIndex, datetime
```

```python
firehose_path = '/cellar/data/TCGA/Firehose__2012_10_24/'
stddata_path = firehose_path + 'stddata'
cancers = os.listdir(stddata_path)
clin_file = lambda c: firehose_path + '/'.join(['stddata', c,'Clinical', c]) + '.clin.merged.txt' 
```

```python
clin = dict((c,read_table(clin_file(c), index_col=0)) for c in cancers
             if os.path.isfile(clin_file(c)))
common = set.intersection(*map(lambda v: set(v.index), clin.values()))
```

```python
keeper_columns = ['bcrpatientbarcode', 'gender', 'daystoinitialpathologicdiagnosis',
                  'daystobirth', 'race', 'icdo3histology', 'vitalstatus', 
                  'patientid', 'tissuesourcesite','daystodeath',
                  'ageatinitialpathologicdiagnosis', 'daystolastfollowup',
                  'ethnicity', 'drugs']
keeper_columns = map(lambda s: 'patient.' + s, keeper_columns)
```

```python
from pandas import Panel, DataFrame, Series
meta_data = {}
```

```python
for c,br in clin.iteritems():
    meta = br.ix[keeper_columns]
    meta = DataFrame(meta.T.as_matrix(), columns=map(lambda s: s.split('.')[-1], 
                     meta.index), index=meta.ix['patient.bcrpatientbarcode'])
    #meta = meta.dropna(how='all', axis=[0,1])
    assert sum(meta['daystoinitialpathologicdiagnosis'] == 0) == 0
    meta_data[c] = meta
```

```python
cols = [c for c, vals in br.iterrows() if (sum(notnull(vals)) > len(vals)*.5)
        and len(set(vals.dropna())) > 1 and ('admin' not in c) and (c not in common)
        and ('patient.drugs.drug' not in c) and ('patient.followups.followup' not in c)
        and ('patient.radiations' not in c)]
```

```python
c  = DataFrame(index=br.columns)
c['metastasis'] = br.ix['patient.distantmetastasispathologicspread']
c['hemoglobin'] = br.ix['patient.hemoglobinresult'].map({'low':-1.,'normal':0, 'elevated':1}) 
c['tumor_grade'] = br.ix['patient.neoplasmhistologicgrade']
c['tumor_status'] = br.ix['patient.personneoplasmcancerstatus']
c['tumor_spread'] = br.ix['patient.primarytumorpathologicspread'].str.slice(stop=2)
c['calcium_level'] = br.ix['patient.serumcalciumresult'].map({'low':-1.,'normal':0, 'elevated':1}) 
c['tumor_stage'] = br.ix['patient.tumorstage']
c['white_cell_count'] = br.ix['patient.whitecellcountresult'].map({'low':-1.,'normal':0, 'elevated':1}) 
c = c.rename(index=br.ix['patient.bcrpatientbarcode'])
```
