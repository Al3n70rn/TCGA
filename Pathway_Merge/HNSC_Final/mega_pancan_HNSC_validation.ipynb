<h1 class="alert alert-info">Pan-Cancer Cohort <small>  <i class="icon-resize-full"></i>  Followup of HNSCC subtypes</small></h1>

* Here we are probing all of the TCGA data that we can get our hands on to support our findings from the HNSCC cohort.
* We do our best to indentify and remove confounders that we expect to impact prognosis independent of the patients' genomic state.
* Due to the fact that the barcodes are unique to each patient, we take the strategy of just getting all of the data into big Series or DataFrames, and then using the __codes__ vector to track the tumor type.

## Read in all of the data for each cancer

```python
cancers = {}
clinical = {}
mut = {}
cn = {}
rna = {}
mirna = {}
rppa = {}
for cancer in run.cancers:
    if cancer == 'COADREAD':
        continue
    try:
        c = run.load_cancer(cancer)
        clinical[cancer] = c.load_clinical()
        mut[cancer] = c.load_data('Mutation')
        cn[cancer] = c.load_data('CN_broad')
        rna[cancer] = c.load_data('mRNASeq')
        mirna[cancer] = c.load_data('miRNASeq')
        rppa[cancer] = c.load_data('RPPA')
        cancers[cancer] = c
    except:
        do_nothing = True
```

## Update to most recent clinical data

```python
from Data.ProcessClinical import format_survival_from_data_portal
for cancer,clin in clinical.iteritems():
    try:
        path  = '/cellar/users/agross/Data/TCGA/Followup/{}/'.format(cancer)
        survival, timeline = format_survival_from_data_portal(path, cancer)
        clin.survival = survival
        clin.artificially_censor(5)
        clinical[cancer] = clin
    except:
        print cancer
        do_nothing = True
```

## Format combined survival objects

```python
surv = pd.DataFrame({c: v.survival.survival for c,v in clinical.iteritems()})
surv = surv.stack()

codes = pd.Series({s[0]: s[1] for s in surv.ix[:,'days'].index})
codes = codes.groupby(level=0).first()
codes.name = 'cancer'

surv.index = surv.index.droplevel(2)
surv = surv.groupby(level=[0,1]).first()

surv_5y = pd.DataFrame({c: v.survival.survival_5y for c,v in clinical.iteritems()})
surv_5y = surv_5y.stack()
surv_5y.index = surv_5y.index.droplevel(2)
surv_5y = surv_5y.groupby(level=[0,1]).first()

bad = ((surv_5y.unstack()['days'] < 8) * (surv_5y.unstack()['event'] == 1)) == False
surv_5y = surv_5y.unstack().ix[bad].stack()
```

```python
for c in clinical.values():
    c = c.artificially_censor(3)
surv_3y = pd.DataFrame({c: v.survival.survival_3y for c,v in clinical.iteritems()})
surv_3y = surv_3y.stack()
surv_3y.index = surv_3y.index.droplevel(2)
surv_3y = surv_3y.groupby(level=[0,1]).first()
```

## Read in most recent mutation data from latest MAFs 


* These are processed in __get_updated_clinical.ipynb__

```python
all_mut = pd.read_csv('/cellar/users/agross/Data/TCGA/New_MAFs/meta.csv', index_col=0)
all_mut = all_mut[all_mut.Tumor_Sample_Barcode.apply(lambda s: s[13:16]) == '01A']
all_mut.Tumor_Sample_Barcode = all_mut.Tumor_Sample_Barcode.map(lambda s: s[:12])
all_mut = all_mut.groupby(['Tumor_Sample_Barcode', 'Hugo_Symbol']).size()
all_mut = all_mut.unstack().T.fillna(0)
```

## Process mutation data into one big matrix 

* We rely on the __codes__ variable to map patients to cancer types 
* We have a little more confidence in data processed through Firehose, therefore we use this data first and fill in mutation profiles from the Data Portal MAFs only when this data is missing.

```python
hit_mat = pd.concat([m.df.unstack() for c,m in mut.iteritems()], axis=1).fillna(0)
hit_mat = hit_mat.groupby(axis=1, level=0).first()
hit_mat = pd.concat([hit_mat, all_mut], axis=1)
hit_mat = hit_mat.groupby(axis=1, level=0).first()
```

## Read in copy number gene by patient matricies 

* Different cancers have different significant copy-number breakpoints 
* For this analysis we just use the gene by patient matrix and estimate the region level values

```python
cn_all = {}
for c in cn.keys():
    cn_all[c] = FH.get_gistic_gene_matrix(run.data_path, c)
cn_all = pd.concat(cn_all.values(), axis=1)
cn_all = cn_all.groupby(axis=1, level=0).first()
```

## Create big miRNA by patient matrix 

```python
mir_mat = pd.concat([mirna[c].df for c in mirna], axis=1)
mir_mat = mir_mat.groupby(level=[0,1], axis=1).first()
```

```python
rna_mat = pd.concat([rna[c].df for c in rna], axis=1)
rna_mat = rna_mat.groupby(level=[0,1], axis=1).first()
```

## Stage

* We try pathologic stage first and then clinical stage 
* I'm just using the broad stage number rather than subcategories 
* Many patients are missing stage information

```python
s = {}
for can,c in clinical.iteritems():
    if 'pathologicstage' in c.stage:
        s[can] = c.stage.pathologicstage
    elif 'clinicalstage' in c.stage:
        s[can] = c.stage.clinicalstage
stage = pd.concat(s.values())
stage = stage.groupby(level=0).first()
stage = stage.dropna().map(lambda s: s.replace('a','').replace('b','').replace('c',''))
stage = stage.dropna().map(lambda s: s.replace('1','').replace('2',''))
stage = stage.ix[age.index].fillna('M')
stage = stage.replace(['stge x','stge tis'], nan).dropna()
stage.name = 'stage'
```

```python
survival_and_stats(stage.ix[true_index(nn)].dropna(), surv_5y)
```

## M-Stage 

* Overlaps with stage but slightly different 
* In HNSCC stage IVa is M0

```python
mets = pd.concat([c.stage.pathologicm for can,c in clinical.iteritems() if 'pathologicm' in c.stage])
mets = mets.groupby(level=0).first()
mets = mets.dropna().map(lambda s: s.replace('a','').replace('b','').replace('c',''))
mets = mets.replace(['stge x','stge tis'], nan).dropna()
mets = mets=='m1'
mets.name = 'metastasis'
survival_and_stats(mets, surv_5y)
```

```python
survival_and_stats(combine(mets, stage=='stge iv'), surv_5y)
```

# Form final Pan-Cancer working set of patients

```python
filters = pd.concat([age > 85, mets, r_status=='r2', stage=='stge iv', nn==False], 1)
keepers = true_index((filters > 0).sum(1) == 0)
keepers = keepers.intersection(surv_5y.unstack().index)
keepers = keepers.intersection(hit_mat.columns).intersection(cn_all.columns)
```
